```python
def recommendation_user(user_id,k,n=10):
    
    s = reco_corr.loc[user_id].sort_values(ascending = False)[1:k+1].index
    
    temp = pd.DataFrame(columns = ['user_id','item_id','rating']) # concat 쓰기 위해 
    
    for j,i in enumerate(s):
       
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        
        k = data[(data.item_id.isin(t)) & (data.rating == 5) & (data.user_id == i)]
        
        k.rating = k.rating * (1-0.05*j)
        
        temp = pd.concat([temp,k])
    
    else:
        
        temp = temp.rename(columns = {'item_id':'movie_id'})
        temp = temp.merge(item[['movie_id','movie_title']]).drop_duplicates() # merge : 관련 정보끼리 inner join
        temp = temp.groupby('movie_title').mean().sort_values('rating',ascending=False)[:n]

        return [_ for _ in temp.index]
```
