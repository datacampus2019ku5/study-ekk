```python
# 알고리즘 불러오기  # KNeighborsClassifier # KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 이름: Camel방식 -> dtype : 클래스. => 인스턴스화.
knn = KNeighborsClassifier()
# 하이퍼 파라미터: 파라미터가 성능에 영향을 미침.
# KNeighborsClassifier(
#     n_neighbors=5,weights='uniform', algorithm='auto', leaf_size=30,
#     p=2, metric='minkowski', metric_params=None, n_jobs=None, **kwargs,)
# n_jobs : 몇개의 콜을 가지고 학습을 시킬 것인가
```

```python
# 컬럼 추출법
# iris[iris.columns[:-1]]
# iris.iloc[:,:-1]
```

```python
# 피터노빅 아저씨 -> "데이터의 양이 많으면 알고리즘별 성능차이가 크게 나지 않는다."
# (하지만, 세상엔 좋은 데이터가 없기 때문에 알고리즘이 중요해졌다.)

#학습시키기 (Training / Fitting)   #Training_data, target_data
#knn.fit(X, y)     # 관례상 행렬은 대문자 쓴다. 벡터는 소문자. # 4줄.

knn.fit(iris[iris.columns[:-1]], iris[iris.columns[-1]])   # 타겟제외 나머지, 타겟
```

```python
knn2 = KNeighborsClassifier()
knn2.fit(iris[iris.columns[:-1]], iris[iris.columns[-1]])
vars(knn2)
```
