```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips
#tips의 내용을 바꿀 것이다.(+1하면 broadcasting으로 바꿀 수 있었다.)
```

# map  : Series

```python
#성별을 남/여를 1/2로 바꾼다 
tips.sex.map({'Male':1,'Female':2})    #map : Series에만 쓴다.
#함수냐 딕셔너리냐 시리즈냐.... 함수만 쓴다.

#함수로 쓸 때
tips.sex.map(lambda x:1 if x=='Male' else 2)
```

## pandas = 데이터 랭글링 툴.(데이터 가져오고 전처리하고 간단한 그래프 그리고.) + 어그리게이션 분석.
## => 쉽게 바꿀 수 있어야 한다.

# apply : Series + Dataframe

```python
tips.apply(lambda x: x['total_bill']+x['tip'], axis=1)
```

```python
tips[['total_bill','tip']].apply(lambda x:sum(x))   #axis=0
```

# applymap

```python
tips[['total_bill','tip']].applymap(lambda x:x+1)
# 전체 값 각각에 +1
```

# 남자 vs 여자 팁 누가 많이 주나(비율)

```python
tips['ratio'] = tips.apply(lambda x:x['tip']/x['total_bill'], axis=1)
```
