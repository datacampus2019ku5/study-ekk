```python
data = load_boston()

boston = pd.DataFrame(data.data, columns=data.feature_names)

boston_target = pd.DataFrame(data.target, columns=['target'])

boston = pd.concat([boston,boston_target],axis=1)
```

```python
from sklearn.datasets import load_iris

datai = load_iris()

iris = pd.DataFrame(datai.data, columns=datai.feature_names)

iris_target = pd.DataFrame(datai.target, columns=['target'])

iris = pd.concat([iris,iris_target],axis=1)
```
